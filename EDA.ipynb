```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv(f"/Users/alexanderbrost/Desktop/root/py/LoanDataset_clean.csv")
```

```python
# bin income into groups

bins = [0, 25000, 50000, 75000, float('inf')]
labels = ["0-25k", "25-50k", "50-75k", "75k+"]
df['income_group'] = pd.cut(df['customer_income'], bins=bins, labels=labels)

# numericize loan status (DEF = 1)
df['Current_loan_status'] = df['Current_loan_status'].replace({'DEFAULT': 1, 'NO DEFAULT': 0})

# calculate default rate by income group
default_rate = df.groupby('income_group')['Current_loan_status'].mean()

# plot income bin vs default rate
plt.figure(figsize=(8, 6))
default_rate.plot(kind='bar', color='skyblue', edgecolor='black')

plt.xlabel('Income Group')
plt.ylabel('Default Rate')
plt.title('Default Rate by Income Group')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# look at loan grade versus default rate

default_rate_profession = df.groupby('loan_grade')['Current_loan_status'].mean()

# plot income bin vs default rate
plt.figure(figsize=(8, 6))
default_rate_profession.plot(kind='bar', color='skyblue', edgecolor='black')

plt.xlabel('Loan Grade')
plt.ylabel('Default Rate')
plt.title('Default Rate by Loan Grade')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# look at loan intent versus default rate

default_rate_intent = df.groupby('loan_intent')['Current_loan_status'].mean()

# plot income bin vs default rate
plt.figure(figsize=(8, 6))
default_rate_intent.plot(kind='bar', color='skyblue', edgecolor='black')

plt.xlabel('Loan Intent')
plt.ylabel('Default Rate')
plt.title('Default Rate by Loan Intent')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# loan amount versus default rate

bins = [0, 10000, 20000, 30000, float('inf')]
labels = ["0-10k", "10-20k", "20-30k", "30k+"]
df['loan_amnt'] = pd.cut(df['loan_amnt'], bins=bins, labels=labels)

default_rate_loan = df.groupby('loan_amnt')['Current_loan_status'].mean()

plt.figure(figsize=(8, 6))
default_rate_loan.plot(kind='bar', color='skyblue', edgecolor='black')

plt.xlabel('Loan Amount')
plt.ylabel('Default Rate')
plt.title('Default Rate by Loan Amount')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# interest rate versus default rate

bins = [0, 6, 12, 24, float('inf')]
labels = ["0-6%", "6-12%", "12-24%", "24%+"]
df['loan_int_rate'] = pd.cut(df['loan_int_rate'], bins=bins, labels=labels)

default_rate_interest = df.groupby('loan_int_rate')['Current_loan_status'].mean()

plt.figure(figsize=(8, 6))
default_rate_interest.plot(kind='bar', color='skyblue', edgecolor='black')

plt.xlabel('Loan Interest Rate')
plt.ylabel('Default Rate')
plt.title('Default Rate by Loan Interest Rate')

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
