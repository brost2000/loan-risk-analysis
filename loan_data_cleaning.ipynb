```python
import pandas as pd

df = pd.read_csv(f"/Users/alexanderbrost/Desktop/root/py/banking_data/LoanDataset - LoansDatasest.csv")
```

```python
# take basic look at dataset and basic descriptive info

print(
    df.head(),
    df.tail(),
    df.info(),
    df.describe(),
    df.isnull().any()
)
```

```python
# convert currency amounts to float

df['loan_amnt'] = df['loan_amnt'].replace('[£,]', '', regex=True).astype(float)
df['customer_income'] = df['customer_income'].replace('[£,]', '', regex=True).astype(float)
```

```python
# historical default is Y/N - convert to binary

df['historical_default'] = df['historical_default'].map({'Y': 1, 'N': 0})
```

```python
# drop null rows in important analysis colummns
clean_df = df.dropna(subset = [
    'customer_income', 
    'loan_amnt', 
    'loan_int_rate', 
    'Current_loan_status',
    'employment_duration',
    'cred_hist_length',
    'loan_intent',
    'loan_grade'
])
```

```python
missing_summary = df.isnull().sum().sort_values(ascending = False)
print("Missing values per column:\n", missing_summary)
```

```python
clean_df.head()
clean_df.to_csv(f"LoanDataset - LoansDatasest_clean.csv")
```
